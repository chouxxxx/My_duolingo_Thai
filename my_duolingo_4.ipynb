```python
import numpy as np
import pandas as pd
import random
from IPython.display import clear_output, display, Markdown

def pick(ser):
    opts = set(ser.values)
    try:
        opts.remove(np.nan)
    except KeyError:
        pass
    opts = list(opts)
    wghts = [ser.value_counts()[opts[i]] for i in range(len(opts))]
    return random.choices(opts, weights=wghts, k=1)[0]

vertical = 0

right = '&emsp;&emsp;&emsp;&emsp;<font size="8" color="#40D040"><strong>{:>42}'\
        '</strong></font>'.format("CORRECT!")
wrong = '&emsp;&emsp;<font size="6" color="#D04040">{:>42}'\
        '</font>'.format("Incorrect.")
t1_0 = '<font size="3">'
t1_1 = '<font size="6">'
t1_2 = '<strong>'
t2 = '</strong>'
tS = '&emsp;'
t3 = '</font>'
t4 = '<font size="5">The answer was:</font>&emsp;<font size="6">'
t5 = '<font size="5">It corresponds to:</font>&emsp;<font size="6">'

data = pd.read_csv("EN_TH.csv")

alphs = ["all", "g", "k", "p", "s", "t"]
weights_alphs = [1, 2.5, 1.5, 2, 1, 2.5]

def display_results(ans):
    if options[ans] == correct_translation:
        display(Markdown(right))
        if dic_pick == "word":
            display(Markdown(ict))
    else:
        display(Markdown(wrong))
        display(Markdown(tru))
        if dic_pick == "word":
            display(Markdown(ict))

while True:
    # Select a random exercise
    transes = ['Karaoké', 'Thaï'] # Sera overwritten si dic_pick == "word"
    dic_pick = pick(data.dic_type)
    subdf = data[data.dic_type == dic_pick]
    
    if dic_pick == "word":
        L_pick = pick(subdf.subtype)
        subdf = subdf[subdf.subtype == L_pick]
        t_pick = pick(subdf.ssubtype)
        subdf = subdf[subdf.ssubtype == t_pick]
        #if len(data[
        #    (data.dic_type == dic_pick)
        #    & (data.subtype == L_pick)
        #    & (data.ssubtype == t_pick)
        #]) > 3:
        if t_pick == "TH-EN":
            transes = ['English', 'Thaï']
        elif t_pick == "KR-EN":
            transes = ['English', 'Karaoké']
        else:
            print("An unidentified dictionnary was picked.")
            pass
    elif dic_pick == "alph":
        alpha = random.choices(alphs, weights=weights_alphs, k=1)[0]
        #print("Sous-alphabet {}".format(alpha))
        subdf = subdf[subdf.subtype.str.contains(alpha)]
    else:
        pass
        # Here dic_pick is either digit or wowel
        # Both are KR-TH
    
    # Select a random source language
    src_lang = random.choice(transes)
    dest_lang = random.sample([l for l in transes if l != src_lang], 1)[0]

    # Select a random row from the dataframe
    random_row = subdf.sample(1)
    original_word = random_row.iloc[0][src_lang]
    correct_translation = random_row.iloc[0][dest_lang]
    
    if dic_pick == "word":
        otr_lang = [l for l in ["English", "Karaoké", "Thaï"]\
                    if (l != src_lang) and (l != dest_lang)][0]
        otr_translation = random_row.iloc[0][otr_lang]

    # Select three random incorrect words from the destination language column
    # incorrect_translations = df.loc[~random_row, dest_lang].sample(3).tolist()
    incorrect_translations = subdf.loc[
        subdf[dest_lang] != correct_translation,
        dest_lang,
    ].sample(3).tolist()

    # Add the correct translation to the list of incorrect translations
    options = incorrect_translations + [correct_translation]
    random.shuffle(options)

    # Print the question and the multiple choices
    display(Markdown("{}{}{}{}{}{}{}{}{}{}{}{}{}".format(
        '<font size="2">',
        "Translate the following word from ",
        '<strong>',
        src_lang,
        '</strong>',
        " to ",
        '<strong>',
        dest_lang,
        '</strong>',
        ": ",
        '</font><font size="6">&emsp;',
        original_word,
        '</font>',
    )))
    if vertical:
        display(Markdown("{}{}{}{}{}{}{}".format(t1_1, t1_2, "W. ", tS, t2, options[0], t3)))
        display(Markdown("{}{}{}{}{}{}{}".format(t1_1, t1_2, "X. ", tS, t2, options[1], t3)))
        display(Markdown("{}{}{}{}{}{}{}".format(t1_1, t1_2, "C. ", tS, t2, options[2], t3)))
        display(Markdown("{}{}{}{}{}{}{}".format(t1_1, t1_2, "V. ", tS, t2, options[3], t3)))
    else:
        display(Markdown(
            "<table style='border:0px solid black;margin-left:auto;margin-right:auto;padding:2px;border-spacing:20px;'>"\
            "    <thead>"\
            "        <tr>"\
            "            <th>{}{}{}{}{}</th>"\
            "            <th>{}{}{}{}{}</th>"\
            "            <th>{}{}{}{}{}</th>"\
            "            <th>{}{}{}{}{}</th>"\
            "        </tr>"\
            "    </thead>"\
            "    <tbody>"\
            "        <tr>"\
            "            <td style='text-align:center'>{}{}{}{}{}</td>"\
            "            <td style='text-align:center'>{}{}{}{}{}</td>"\
            "            <td style='text-align:center'>{}{}{}{}{}</td>"\
            "            <td style='text-align:center'>{}{}{}{}{}</td>"\
            "        </tr>"\
            "    </tbody>"\
            "</table>".format(
                t1_0, t1_2, "W", t2, t3,
                t1_0, t1_2, "X", t2, t3,
                t1_0, t1_2, "C", t2, t3,
                t1_0, t1_2, "V", t2, t3,
                tS, t1_1, options[0], t3, tS,
                tS, t1_1, options[1], t3, tS,
                tS, t1_1, options[2], t3, tS,
                tS, t1_1, options[3], t3, tS,
            )))
    print("P. Quit the program.")
    print("")
    print("")

    # Get the user's answer
    answer = input("Enter the letter of your answer: ").upper()
    clear_output()

    tru = t4 + correct_translation + t3
    if dic_pick == "word":
        ict = t5 + otr_translation + t3
    # Check if the answer is correct
    if answer == 'W':
        display_results(0)
    elif answer == 'X':
        display_results(1)
    elif answer == 'C':
        display_results(2)
    elif answer == 'V':
        display_results(3)
    elif answer == 'P':
        break
    else:
        print("Invalid input.")
    print("")
```
